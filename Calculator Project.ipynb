```python
from tkinter import Tk, Entry, Button

class Calculator:
    def __init__(self, window):
        self.window = window
        window.title("Calculator")

        self.entry = Entry(window, width = 50, borderwidth = 15,bg='black', fg='orange')
        self.entry.grid(row = 0, column = 0, columnspan = 4, padx = 10, pady = 10)

        keys = [
            '7', '8', '9', '+',
            
            '4', '5', '6', '-',
            
            '1', '2', '3', '*',
            
            '0', '.', '=', '/'
               ]

        general_bg_color = 'orange'
        row_val = 1
        col_val = 0
        for key in keys:
            if key in ['*', '/', '+', '-', '=']:
                fg_color = 'blue'
            else:
                fg_color = 'orange'
                
            Button(window, text=key, padx=5, pady=10, width=10, height=2, bg='black', fg='orange',
                   command=lambda k=key: self.entry.insert('end', k) if k != '=' else self.calculate()).grid(row = row_val,
                                                                                                             column=col_val)
            
            col_val += 1
            if col_val > 3:
                col_val = 0
                row_val += 1

    def calculate(self):
        try:
            result = eval(self.entry.get())
            self.entry.delete(0, 'end')
            self.entry.insert('end', str(result))
            
        except Exception as error:
            self.entry.delete(0, 'end')
            self.entry.insert('end', 'ERROR!')

window = Tk()

calculator = Calculator(window)

window.mainloop()

            









```
